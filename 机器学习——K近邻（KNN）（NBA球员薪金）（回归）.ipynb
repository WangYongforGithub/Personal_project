```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier  # KNN 分类
from sklearn.neighbors import KNeighborsRegressor   # KNN 回归

%config InlineBackend.figure_format = 'svg'

plt.rcParams['font.sans-serif'] = ['Kaiti']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('NBA球员数据.xlsx')
df.head()
```

# 整理数据

```python
data = df[['身高','体重','年龄','球龄','上场次数','场均时间','进攻能力','防守能力','球员薪金']]
data.head()
```

```python
x = data[['身高','体重','年龄','球龄','上场次数','场均时间','进攻能力','防守能力']]
y = data['球员薪金']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
knn_clf = KNeighborsRegressor(n_neighbors=10, p=2) 
```

```python
knn_clf.fit(x_train, y_train)
```

```python
sal = knn_clf.predict(x_test)
```

```python
# knn_clf.score(x_test, y_test)
```

```python
plt.plot(range(1,121),sal)
plt.plot(range(1,121), y_test)
plt.legend(['sal','y_test'])
plt.show()
```

```python
print(np.std(sal), np.std(y_test))
```
